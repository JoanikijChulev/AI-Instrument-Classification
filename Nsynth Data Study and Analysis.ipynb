## Sample Rate Consistency

```python
# List of your datasets
datasets = [df_train, df_valid, df_test]  # Assuming you have these datasets

# Extracting unique sample rates from each dataset
unique_sample_rates = []

for dataset in datasets:
    unique_sample_rates.append(dataset['sample_rate'].unique())

# Checking if there's only one unique sample rate across all datasets
if len(set(tuple(sample_rate) for sample_rate in unique_sample_rates)) == 1:
    sample_rate = unique_sample_rates[0][0]
    print("Sample rate is consistent across all datasets.")
    print("Sample rate:", sample_rate)
    
    # Plotting a bar chart
    plt.figure(figsize=(8, 5))
    bar_colors = ['skyblue', 'lightgreen', 'lightcoral']
    plt.bar(range(len(datasets)), [sample_rate]*len(datasets), color=bar_colors, alpha=0.7)
    plt.xticks(range(len(datasets)), ['Training', 'Validation', 'Test'], fontsize=12)
    plt.yticks(fontsize=12)
    plt.title("Sample Rate Consistency Across Datasets", fontsize=16)
    plt.xlabel("Dataset Types", fontsize=16)
    plt.ylabel("Sample Rate", fontsize=14)
    plt.ylim(min(sample_rate - 1000, 0), sample_rate + 1000)  # Adjust y-axis limit for better visualization
    plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding grid for better visualization
    plt.tight_layout()
    plt.show()
else:
    print("Sample rates are different across datasets.")
```
